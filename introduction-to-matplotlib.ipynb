## Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for plotting in Jupyter

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # (width, height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
         xlabel="x-axis",
            ylabel="y-axis") # add a title, xlabel, ylabel

# 5. Save & show (you save the whole figure)
fig.savefig("sample-plot.png")
```

# Two options for subplots

```python
# subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, # 2 rows
                                            ncols=2, # 2 columns
                                            figsize=(10, 5)) # create 2x2 figure

# plot to each different axis
ax1.plot(x, x/2); # top left
ax2.scatter(np.random.random(10), np.random.random(10)); # top right
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values()); # bottom left
ax4.hist(np.random.randn(1000)); # bottom right
```

```python
# subplot option 2
fig, ax = plt.subplots(nrows=2, # 2 rows
                        ncols=2, # 2 columns
                        figsize=(10, 5)) # create 2x2 figure

# plot to each different index

ax[0, 0].plot(x, x/2); # top left, column 0, row 0
ax[0, 1].scatter(np.random.random(10), np.random.random(10)); # top right, column 1, row 0
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values()); # bottom left, column 0, row 1
ax[1, 1].hist(np.random.randn(1000)); # bottom right, column 1, row 1
```

## Which one should you use? (pyplot vs matplotlib OO method?)

* when plotting something quickly, okay to use pyplot method
* when plotting something more advanced, use OO method

```python
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease["age"]>50] # get all rows where age is greater than 50
over_50
```

```python
# pyplot method
over_50.plot(kind="scatter", # plot scatter graph
                x="age", # x axis is age
                y="chol", # y axis is chol
                c="target"); # c is color, color is target column
# the target column is 1 if the person has heart disease, 0 if not
```

```python
# OO method mixed with pyplot method
fig, ax=plt.subplots(figsize=(10, 6)) # get a figure and axes

over_50.plot(kind="scatter", # plot scatter graph
                x="age", # x axis is age
                y="chol", # y axis is chol
                c="target", # c is color, color is target column
                ax=ax); # plot to ax, which is the axes we created

ax.set_xlim([45, 100]); # set x axis limits
```

```python
## OO method from scratch
fig, ax=plt.subplots(figsize=(10, 6)) # get a figure and axes, set size (width, height)

# plot the data
scatter=ax.scatter(x=over_50["age"], # x axis is age
                    y=over_50["chol"], # y axis is chol
                    c=over_50["target"]); # c is color, color is target column

# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels", # add title
        xlabel="Age", # add x axis label
        ylabel="Cholesterol"); # add y axis label

# add a legend
ax.legend(*scatter.legend_elements(), title="Target");
 # add a legend, *scatter.legend_elements() 
 # gets the legend elements from the scatter plot
 # title is the title of the legend

# add a horizontal line
ax.axhline(over_50["chol"].mean(), # add a horizontal line
            linestyle="--"); # add a horizontal line, linestyle is dashed

```

```python
over_50.head()
```

```python
# subplot of chol, age, thalach

# setup plot, with 2 rows, 1 column, share x axis
# create a figure and axes
fig, (ax0, ax1)=plt.subplots(nrows=2, # 2 rows
                                ncols=1, # 1 column
                                figsize=(10, 10), # set size (width, height)
                                sharex=True) # share x axis

# add data to ax0
scatter=ax0.scatter(x=over_50["age"], # x axis is age
                    y=over_50["chol"], # y axis is chol
                    c=over_50["target"]); # c is color, color is target column

# customize ax0 plot, located at row 0, column 0
ax0.set(title="Heart Disease and Cholesterol Levels", # add title
        ylabel="Cholesterol"); # add y axis label

# add a legend
ax0.legend(*scatter.legend_elements(), title="Target");
    # add a legend, *scatter.legend_elements()

# add a horizontal/mean line
ax0.axhline(over_50["chol"].mean(), # add a horizontal line
            linestyle="--"); # add a horizontal line, linestyle is dashed



# add data to ax1
scatter=ax1.scatter(x=over_50["age"], # x axis is age
                    y=over_50["thalach"], # y axis is chol
                    c=over_50["target"]); # c is color, color is target column

# customize ax1 plot, located at row 0, column 0
ax1.set(title="Heart Disease and thalach Levels", # add title
        ylabel="thalach"); # add y axis label

# add a legend
ax1.legend(*scatter.legend_elements(), title="Target");
    # add a legend, *scatter.legend_elements()

# add a horizontal/mean line
ax1.axhline(over_50["thalach"].mean(), # add a horizontal line
            linestyle="--"); # add a horizontal line, linestyle is dashed


# add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

fig.savefig("heart-disease-analysis-plot.png"); # save the figure

```

```python
# set the style
plt.style.use("seaborn-whitegrid") # use seaborn-whitegrid style

## OO method from scratch
fig, ax=plt.subplots(figsize=(10, 6)) # get a figure and axes, set size (width, height)

# plot the data
scatter=ax.scatter(x=over_50["age"], # x axis is age
                    y=over_50["chol"], # y axis is chol
                    c=over_50["target"], # c is color, color is target column
                    cmap="winter"); # change the color scheme

# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels", # add title
        xlabel="Age", # add x axis label
        ylabel="Cholesterol"); # add y axis label

# add a legend
ax.legend(*scatter.legend_elements(), title="Target");
 # add a legend, *scatter.legend_elements() 
 # gets the legend elements from the scatter plot
 # title is the title of the legend

# add a horizontal line
ax.axhline(over_50["chol"].mean(), # add a horizontal line
            linestyle="--"); # add a horizontal line, linestyle is dashed
```

```python
## customizing the y and x axis limitations

# subplot of chol, age, thalach

# setup plot, with 2 rows, 1 column, share x axis
# create a figure and axes
fig, (ax0, ax1)=plt.subplots(nrows=2, # 2 rows
                                ncols=1, # 1 column
                                figsize=(10, 10), # set size (width, height)
                                sharex=True) # share x axis

# add data to ax0
scatter=ax0.scatter(x=over_50["age"], # x axis is age
                    y=over_50["chol"], # y axis is chol
                    c=over_50["target"], # c is color, color is target column
                    cmap="winter"); # change the color scheme

ax0.set_xlim([50, 80]) # set x axis limits

# customize ax0 plot, located at row 0, column 0
ax0.set(title="Heart Disease and Cholesterol Levels", # add title
        ylabel="Cholesterol"); # add y axis label

# add a legend
ax0.legend(*scatter.legend_elements(), title="Target");
    # add a legend, *scatter.legend_elements()

# add a horizontal/mean line
ax0.axhline(over_50["chol"].mean(), # add a horizontal line
            linestyle="--"); # add a horizontal line, linestyle is dashed



# add data to ax1
scatter=ax1.scatter(x=over_50["age"], # x axis is age
                    y=over_50["thalach"], # y axis is chol
                    c=over_50["target"], # c is color, color is target column
                    cmap="winter"); # change the color scheme

ax1.set_xlim([50, 80]) # set x axis limits
ax1.set_ylim([60, 200]) # set y axis limits

# customize ax1 plot, located at row 0, column 0
ax1.set(title="Heart Disease and thalach Levels", # add title
        xlabel="Age", # add x axis label
        ylabel="thalach"); # add y axis label

# add a legend
ax1.legend(*scatter.legend_elements(), title="Target");
    # add a legend, *scatter.legend_elements()

# add a horizontal/mean line
ax1.axhline(over_50["thalach"].mean(), # add a horizontal line
            linestyle="--"); # add a horizontal line, linestyle is dashed


# add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

fig.savefig("heart-disease-analysis-plot.png"); # save the figure
```
